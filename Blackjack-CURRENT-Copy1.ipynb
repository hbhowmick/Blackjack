# Blackjack

1) Win condition (higher score) <br>
2) Lose condition (someone busts) <br>
3) OOP included (use of classes) <br>
4) Ability to 'hit' or 'stay' during hand play <br>
5) If the total is the same it's a 'push' (a tie where no one wins) <br>
6) The game should continue to run until the user quits <br>
7) Aces are worth 11 or 1 depending on total score, if there is an ace in the player's hand and they go over 21, then the ace becomes the value of 1 <br>
8) Overall gameplay, dealer hits while under 17, higher score wins <br>

Extra features:

1) Add currency <br>
2) Multiplayer <br>
3) Using real decks of cards (double for loop, one to create the suit, one to create the value of the card (store in tuple) <br>
4) Adding the ability to 'double down' or 'split' <br>
